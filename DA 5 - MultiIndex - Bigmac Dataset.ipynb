## The `.set_names()` Method on MultiIndex

```python
df = pd.read_csv('pandas/bigmac.csv', parse_dates = ['Date'], index_col=['Date','Country'])
df.sort_index(inplace = True)
df.head(3)
```

```python
df.index.set_names(['Date','Location'], inplace=True)
df.head(3)
```

## The `.transpose()` Method

```python
df = pd.read_csv('pandas/bigmac.csv', parse_dates = ['Date'], index_col=['Date','Country'])
df.sort_index(inplace = True)
df.head(3)
```

```python
df = df.transpose()
df.head()
```

```python
df.loc["Price in US Dollars", ("2016-01-01", "Denmark")]
```

## The `pivot` Method

```python
df = pd.read_csv('pandas/salesmen.csv', parse_dates=['Date'])
df['Salesman'] = df['Salesman'].astype('category')
df.head(3)
```

```python
df['Salesman'].value_counts()
```

```python
df.pivot(index='Date', columns='Salesman', values='Revenue').head()
```

```python
len(df.pivot(index='Date', columns='Salesman', values='Revenue'))
```

## The `pivot_table()` Method

```python
df = pd.read_csv('pandas/foods.csv')
df.head()
```

```python
df.pivot_table(values='Spend', index='Gender', aggfunc='mean')
```

```python
df.pivot_table(values='Spend', index='Item', aggfunc='sum')
```

```python
df.pivot_table(values='Spend', index=['Gender', 'Item'], aggfunc='min')
```

```python
df.pivot_table(values='Spend', index=['Gender', 'Item'], columns='City', aggfunc='count')
```

```python
df.pivot_table(values='Spend', index=['Gender', 'Item'], columns=['Frequency', 'City'], aggfunc='max')
```
