## The `.astype()` Method

```python
nba = pd.read_csv("pandas/nba.csv")
nba["Team"].fillna('None', inplace=True)
nba["Number"].fillna(0, inplace=True)
nba["Position"].fillna('None', inplace=True)
nba["Age"].fillna(0, inplace=True)
nba["College"].fillna("No College", inplace=True)
nba["Salary"].fillna(0, inplace=True)
nba.head()
```

```python
nba.dtypes
```

```python
nba["Number"] = nba["Number"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba["Salary"] = nba["Salary"].astype("int")
```

```python
nba.head()
```

```python
nba.dtypes
```

```python
nba["Salary"] = nba["Salary"].astype("float")
```

```python
nba.head(3)
```

## Rank Values with the `.rank()` Method

```python
nba = pd.read_csv("pandas/nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int")
nba.head(3)
```

```python
nba.sort_values("Salary", ascending=False).head(3)
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending=False).astype("int")
nba.head(3)
```

```python
nba.sort_values("Salary", ascending = False)
```
