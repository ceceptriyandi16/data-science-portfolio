## The `.agg()` Method

```python
df_index = pd.read_csv('pandas/fortune1000.csv', index_col='Rank')
df_index.head(3)
```

```python
df_index.groupby('Sector').agg({'Revenue': 'sum',
                               'Profits': 'sum',
                               'Employees': 'mean'})
```

```python
df_index.groupby('Sector').agg({'Revenue': ['sum', 'mean'],
                               'Profits': 'sum',
                               'Employees': 'mean'}).head()
```

```python
pd.set_option('precision', 3)
df_index.groupby('Sector').agg(['size', 'sum', 'min', 'mean', 'max']).head()
```
