## `.apply()` Method on Single Columns

```python
df = pd.read_csv('pandas/jamesbond.csv', index_col='Film')
df.sort_index(inplace=True)
df.head(3)
```

```python
def convert_str(num):
    x = str(num) + ' Millions'
    return x
```

```python
columns=['Box Office','Budget','Bond Actor Salary']
for i in columns:
    df[i] = df[i].apply(convert_str)
```

```python
df.head(3)
```

## `.apply()` Method with Row Values

```python
df = pd.read_csv('pandas/jamesbond.csv', index_col='Film')
df.sort_index(inplace=True)
df.head(3)
```

```python
def movie(col):
    actor=col[1]
    budget=col[4]
    
    if actor == 'David Niven':
        return 'The Best'
    elif actor == 'Roger Moore' and budget > 40:
        return 'Enjoyable'
    else:
        return "I haven't clue"

df.apply(movie, axis=1)
```
