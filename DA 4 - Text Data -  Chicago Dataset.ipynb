## Common String Methods: `.lower()`, `.upper()`, `.title()`, and `.len()`

```python
df = pd.read_csv('pandas/chicago.csv')
df['Department'] = df['Department'].astype('category')
df.head()
```

```python
df['Name'].str.lower().head()
```

```python
df['Name'] = df['Name'].str.title()
df.head()
```

```python
df['Name'] = df['Name'].str.upper()
df.head()
```

```python
df['Position Title'] = df['Position Title'].str.title()
df['Department'] = df['Department'].str.title()
df.head()
```

```python
df['Department'].str.len().head()
```

## `.str.replace()` Method

```python
df = pd.read_csv('pandas/chicago.csv')
df['Department'] = df['Department'].astype('category')
df.head()
```

```python
df['Department'] = df['Department'].str.replace('MGMNT', 'MANAGEMENT')
df['Department'] = df['Department'].astype('category')
df.head()
```

```python
df.info()
```

```python
df['Employee Annual Salary'] = df['Employee Annual Salary'].str.replace('$', '')
df['Employee Annual Salary'] = df['Employee Annual Salary'].astype('float')
df.head()
```

```python
df.info()
```

```python
print('Sum \t: ', df['Employee Annual Salary'].sum())
print('Count \t: ', df['Employee Annual Salary'].count())
print('Max \t: ', df['Employee Annual Salary'].max())
print('Min \t: ', df['Employee Annual Salary'].min())
```

```python
df['Employee Annual Salary'].nlargest()
```

```python
df['Employee Annual Salary'].nsmallest()
```

## Filtering with String Methods

```python
df = pd.read_csv('pandas/chicago.csv').dropna(how='all')
df['Department'] = df['Department'].astype('category')
df.head()
```

```python
df.isnull().sum()
```

```python
df['Position Title'] = df['Position Title'].str.lower()
df.head()
```

```python
df[df['Position Title'].str.lower().str.contains('water')]
```

```python
df[df['Position Title'].str.lower().str.startswith('water')]
```

```python
df[df['Position Title'].str.lower().str.endswith('ist')]
```

## More String Methods: strip, lstrip, rstrip

```python
df = pd.read_csv('pandas/chicago.csv')
df['Department'] = df['Department'].astype('category')
df.head()
```

```python
df['Name'] = df['Name'].str.rstrip().str.lstrip()
df.head()
```

```python
df['Position Title'] = df['Position Title'].str.strip()
df.head()
```

## String Methods on Index and Columns

```python
df = pd.read_csv('pandas/chicago.csv', index_col='Name').dropna(how='all')
df['Department'] = df['Department'].astype('category')
df.head()
```

```python
df.index = df.index.str.strip().str.title()
```

```python
df.head()
```

```python
df.columns = df.columns.str.upper()
```

```python
df.head()
```

## Split Strings by Characters with `.str.split()` Method

```python
df = pd.read_csv('pandas/chicago.csv').dropna(how='all')
df['Department'] = df['Department'].astype('category')
df.head()
```

```python
df['Name'].str.split(',')
```

```python
df['Name'].str.split(',').str.get(0)
```

```python
df['Name'].str.split(',').str.get(0).str.title().value_counts().head()
```

```python
df['Name'].str.split(',').str.get(1).str.strip().str.split(' ').str.get(0).value_counts().head()
```

## The Expand and n Parameters of the `.str.split()` Method

```python
df = pd.read_csv('pandas/chicago.csv').dropna(how='all')
df['Department'] = df['Department'].astype('category')
df.head()
```

```python
df['Name'].str.split(',', expand=True).head()
```

```python
df[['First Name', 'Last Name']] = df['Name'].str.split(',', expand=True)
df.head()
```

```python
df[['First Title Word', 'Remaining Words']] = df['Position Title'].str.split(" ", expand=True, n=1)
df.head()
```
